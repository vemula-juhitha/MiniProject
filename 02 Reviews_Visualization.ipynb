```python
f = open("C:\\Users\\Gopal\\Desktop\\Proj\\dataset\\reviews\\reviews.txt","r")
full_file_corpus=f.read()
print(full_file_corpus)
```

```python
from nltk.tokenize import word_tokenize

try:  # py3
    all_tokens = [t for t in word_tokenize(full_file_corpus)]
except UnicodeDecodeError:  # py27
    all_tokens = [t for t in word_tokenize(full_file_corpus.decode('utf-8'))]

print("Total number of tokens: {}".format(len(all_tokens)))

print(all_tokens)
```

```python
c=input()
l=[i for i in all_tokens if i[0][0].casefold()==c]

print(l)
```

```python
i=input()
print(i,":",all_tokens.count(i))
```

```python
from collections import Counter

total_term_frequency = Counter(all_tokens)

for word, freq in total_term_frequency.most_common(20):
    print("{}\t{}".format(word, freq))
```

```python
from nltk.corpus import stopwords
import string

print(stopwords.words('english'))
print(len(stopwords.words('english')))
print(string.punctuation)
```

```python
from collections import Counter
stop_list = stopwords.words('english') + list(string.punctuation)
st=[token.capitalize() for token in stop_list]
stop_list1=stop_list+st
tokens_no_stop = [token for token in all_tokens
                        if token not in stop_list1]

final_token_string=' '.join([d for d in tokens_no_stop])
print(final_token_string)
total_term_frequency_no_stop = Counter(tokens_no_stop)

for word, freq in total_term_frequency_no_stop.most_common(20):
    print("{}\t{}".format(word, freq))
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
all_tokens_lower = [t.lower() for t in all_tokens]

tokens_normalised = [stemmer.stem(t) for t in all_tokens_lower
                                     if t not in stop_list1]

total_term_frequency_normalised = Counter(tokens_normalised)

for word, freq in total_term_frequency_normalised.most_common(20):
    print("{}\t{}".format(word, freq))
```

```python
from nltk import ngrams

phrases = Counter(ngrams(all_tokens_lower, 2))
for phrase, freq in phrases.most_common(20):
    print("{}\t{}".format(phrase, freq))
```

```python
phrases = Counter(ngrams(all_tokens_lower, 3))
for phrase, freq in phrases.most_common(20):
    print("{}\t{}".format(phrase, freq))
```

```python
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 
  
# Image operation using thresholding 
img = cv2.imread('guitar.jpg') 
  
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
  
ret, thresh = cv2.threshold(gray, 0, 75, 
                            cv2.THRESH_BINARY_INV +
                            cv2.THRESH_OTSU) 
plt.figure(figsize=(12,8))
plt.imshow(thresh,cmap="gray")

plt.axis("off")


plt.show()
```

```python
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 
  
edges = cv2.Canny(thresh,200,200)
plt.figure(figsize=(12,8))
plt.imshow(edges,cmap="gray")
plt.axis("off")


plt.show()
```
